```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data_for_dash_1.csv")
df.head().T
```

```python
df["type"].unique()
```

```python
df["type"] = df["type"].map({
    "Konut" : "Жилье"
})
```

```python
df["sub_type"].unique()
```

```python
df["sub_type"] = df["sub_type"].map({
    "Rezidans": "Резиденция (Жилой комплекс)",
    "Daire": "Квартира",
    "Villa": "Вилла",
    "Müstakil Ev": "Отдельный дом (Частный дом)",
    "Kooperatif": "Кооператив (Кооперативная квартира)",
    "Yazlık": "Летний дом (Дача)",
    "Komple Bina": "Целиком здание (Апарт-отель/Доходный дом)",
    "Prefabrik Ev": "Префабричный дом (Модульный дом)",
    "Köşk / Konak / Yalı": "Особняк / Конак / Ялы (Элитный особняк у моря)",
    "Çiftlik Evi": "Фермерский дом (Усадьба)",
    "Yalı Dairesi": "Квартира в ялы (Апартаменты у моря)",
    "Loft": "Лофт"
})
```

```python
df["listing_type"].unique()
```

```python
df["listing_type"] = df["listing_type"].map({
    1: "Продажа",
    2 : "Аренда",
    3 : "Неизвестно"
})
```

```python
df["heating_type"].unique()
```

```python
df["heating_type"] = df["heating_type"].map({
    "Fancoil": "Фанкойл (система вентиляции/кондиционирования)",
    "Yok": "Нет",
    "Kalorifer (Doğalgaz)": "Калорифер (Природный газ)",
    "Kalorifer (Kömür)": "Калорифер (Уголь)",
    "Kombi (Elektrikli)": "Комби-котёл (Электрический)",
    "Klima": "Кондиционер",
    "Kombi (Doğalgaz)": "Комби-котёл (Природный газ)",
    "Merkezi Sistem (Isı Payı Ölçer)": "Центральное отопление (С тепловым счётчиком)",
    "Merkezi Sistem": "Центральное отопление",
    "Soba (Kömür)": "Печь (Уголь)",
    "Yerden Isıtma": "Тёплый пол",
    "Soba (Doğalgaz)": "Печь (Природный газ)",
    "Güneş Enerjisi": "Солнечная энергия",
    "Kalorifer (Akaryakıt)": "Калорифер (Жидкое топливо)",
    "Jeotermal": "Геотермальное отопление",
    "Kat Kaloriferi": "Этажный калорифер (локальная котельная на этаж)"
})
```

```python
df["price_currency"].unique()
```

```python
df["price_currency"] = df["price_currency"].map({
    "TRY": "TRY (Турецкая лира)",
    "GBP": "GBP (Фунт стерлингов)",
    "EUR": "EUR (Евро)",
    "USD": "USD (Доллар США)",
    float('nan'): "Неизвестно"
})
```
